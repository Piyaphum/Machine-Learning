# Remove Outlier
การคำนวณจากค่าสถิติ IQR เนื่องจากข้อมูลผิดปกติมีค่ามากหรือน้อยกว่าข้อมูลอื่น ๆ นิยมใช้ค่าพิสัยระหว่างควอไทล์ (interquartile range หรือ IQR) ซึ่งเป็นการวัดกระจายข้อมูลด้วยผลต่างหรือระยะห่างระหว่างค่าข้อมูลที่ตำแหน่งควอไทล์ที่ 3 กับค่าข้อมูลที่ตำแหน่งควอไทล์ที่ 1


image.png

```python
import matplotlib.pyplot as plt
# Show boxplots for each numerical columns
for column in ['Age', 'Region_Code', 'Annual_Premium', 'Policy_Sales_Channel', 'Vintage'] :
  plt.figure()
  df.boxplot([column], return_type='axes')
# Show the plot
plt.show()
```

# การคำนวณช่วงค่าปกติ
การคำนวณค่า IQR สามารถทำได้จากการหาค่าควอไทล์ที่ 1 และ 3 ของคอลัมน์ค่าโดยสาร fare แล้วหาผลต่างระหว่างสองค่าดังกล่าว โดยกำหนดให้ค่าผิดปกติแบ่งออกเป็น 2 ช่วงคือ

1.	ค่าที่น้อยกว่าค่าควอไทล์ที่ 1 มากกว่า 1.5 เท่าของ IQR
2.	ค่าที่มากกว่าค่าควอไทล์ที่ 3 มากกว่า 1.5 เท่าของ IQR
คำสั่งจะคัดเลือกเฉพาะรายการข้อมูลที่ไม่ได้อยู่ในช่วงดังกล่าวไว้ในตัวแปร df


```python
cols = ['Age','Policy_Sales_Channel','Annual_Premium']
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
for col in cols:
  print("Remove outliers in ", col)
  df = df[~((df[[col]] < (Q1[col] - 1.5 * IQR[col])) |(df[[col]] > (Q3[col] + 1.5 * IQR[col]))).any(axis=1)]
print(df.shape)
```

# Data Encoding
ขั้นตอนแปลงรหัสข้อมูลเปลี่ยนแปลงข้อมูลจัดกลุ่ม (Categorical data) ให้อยู่ในรูปแบบตัวเลข เพื่อเตรียมพร้อมสำหรับการคำนวณทางคณิตศาสตร์และสถิติ

```python
from sklearn import preprocessing
encode = preprocessing.LabelEncoder()
df['Gender'] = encode.fit_transform(df.Gender.values)
df['Vehicle_Damage'] = encode.fit_transform(df.Vehicle_Damage.values)
df['Vehicle_Age'] = encode.fit_transform(df.Vehicle_Age.values)
df
```

# Data Normalization

# Normalized Data with MinMax Function

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
# Encode age as min-max score
m_scaler = MinMaxScaler()
df[['Age']] = m_scaler.fit_transform(df[['Age']])
df[['Vintage']] = m_scaler.fit_transform(df[['Vintage']])
df[['Annual_Premium']] = m_scaler.fit_transform(df[["Annual_Premium"]])
df[['Region_Code']] = m_scaler.fit_transform(df[["Region_Code"]])
df[['Policy_Sales_Channel']] = m_scaler.fit_transform(df[['Policy_Sales_Channel']])

df
```

# Splitting Data
ตัวแบบการเรียนรู้ด้วยเครื่องมีเป้าหมายเพื่อสร้างการทำนายคำตอบข้อมูลใหม่ที่ไม่เคยเห็นมาก่อนได้อย่างถูกต้อง กระบวนการสร้างแบบจำลองจะแบ่งชุดข้อมูลของคุณออกเป็นสองชุดย่อย คือ
1.	ชุดข้อมูลฝึก (Training set) เป็นชุดข้อมูลฝึกฝนตัวแบบการเรียนรู้ด้วยเครื่อง
2.	ชุดข้อมูลทดสอบ (Test set) เป็นชุดข้อมูลทดสอบประสิทธิภาพของตัวแบบที่ได้จากฝึกฝนของชุดข้อมูลฝึก

  แบบจำลองที่มีประสิทธิภาพในการทำนายผลสูงเป็นเป้าหมายสำคัญในการสร้างแบบจำลอง ทำให้ต้องมีกระบวนการสร้างเรียนรู้และปรับค่าพารามิเตอร์ที่เหมาะสมและประเมินประสิทธิภาพของแบบจำลอง การใช้ข้อมูลเดียวกันในการฝึกฝนเรียนรู้ค่าพารามิเตอร์และทดสอบแบบจำลองไม่ควรกระทำในกระบวนการนี้ เพราะแบบจำลองที่เรียนรู้และทดสอบกับข้อมูลเดียวกันซ้ำจะจดจำคำตอบกับข้อมูลมากเกินไป ทำให้แบบจำลองไม่สามารถทำนายคำตอบกับข้อมูลที่ไม่เคยเห็นมาก่อน (Unseen data) ได้ ปรากฏการณ์นี้เรียกว่า “Overfitting”


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import LeaveOneOut, KFold, StratifiedKFold

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

print(" Display training and test index of data based on Stratifiled 5 folds cross validation")
skf = StratifiedKFold(n_splits = 5)
for train_index, test_index in skf.split(X,y) :
  print("%s %s" %(train_index, test_index))
```
